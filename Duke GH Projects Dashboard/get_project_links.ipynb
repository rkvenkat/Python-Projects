```python
#imports here
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import time
```

```python
driver = webdriver.Chrome('/home/venkat/Downloads/chromedriver_linux64/chromedriver')
driver.get('https://globalhealth.duke.edu/projects?page=100')
```

```python
h1 = driver.find_elements(By.TAG_NAME, 'article')
```

```python
driver.page_source
```

```python
data = []

for i in h1:
    a = i.find_elements(By.TAG_NAME, 'a')
    project_link = a[0].get_attribute('href')
    project_name = a[0].text
    data.append([project_name, project_link])
```

```python
import csv

with open("gh_duke_project_links.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)  
```
